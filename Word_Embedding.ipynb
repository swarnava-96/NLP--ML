# **Word Embedding Techniques using Embedding Layer in Keras**
### Importing the necessary libraries

```python
from tensorflow.keras.preprocessing.text import one_hot
```

### Sentences

```python
sent = ["the glass of milk",
        "the glass of juice",
        "the cup of tea",
        "I am a good boy",
        "I am a good developer",
        "understand the meaning of words",
        "your videos are good"]
```

```python
# Checking out our sentences

sent
```

### Setting Vocabulary size

```python
voc_size = 10000
```

### One Hot Representation

```python
onehot_repr = [one_hot(words, voc_size) for words in sent]
print(onehot_repr)
```

### Word Embedding Representation

```python
# Importing the libraries required for Word Embedding

import numpy as np
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
# Word Embedding

sent_length = 8
embedded_docs = pad_sequences(onehot_repr, padding = "pre", maxlen = sent_length)
print(embedded_docs)
```

```python
# Lets set the number of dimensions for feature representation
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size, 10, input_length = sent_length))
model.compile("adam", "mse")
```

```python
# Lets check the model summary

model.summary()
```

```python
# Lets print the predicted embedded docs

print(model.predict(embedded_docs))
```
